# kNN 
## Метод ближайших соседей
https://www.kaggle.com/code/prashant111/knn-classifier-tutorial
 - **Введение**
    
    В машинном обучении k ближайших соседей или kNN - самый простой из всех алгоритмов машинного обучения. Это непараметрический алгоритм, используемый для задач классификации и регрессии. Непараметрический означает, что для распределения данных не требуется никаких предположений. Таким образом, kNN не требует, чтобы было сделано какое-либо основополагающее предположение. Как в задачах классификации, так и в задачах регрессии входные данные состоят из k ближайших обучающих примеров в пространстве признаков. Выходные данные зависят от того, используется ли kNN для целей классификации или регрессии.

    - В классификации kNN результатом является принадлежность к классу. Данная точка данных классифицируется на основе большинства типов ее соседей. Точка данных присваивается наиболее часто встречающемуся классу среди k ближайших соседей. Обычно k - это небольшое положительное целое число. Если k=1, то точка данных просто присваивается классу этого единственного ближайшего соседа.

    - В регрессии kNN результатом является просто некоторое значение свойства объекта. Это значение является средним из значений k ближайших соседей.

    kNN - это тип обучения на основе экземпляров или ленивого обучения. Ленивое обучение означает, что для генерации модели не требуются какие-либо обучающие точки данных. Все данные обучения будут использованы на этапе тестирования. Это ускоряет обучение, а тестирование замедляет и удорожает стоимость. Таким образом, этап тестирования требует больше времени и ресурсов памяти.

    В kNN соседи берутся из набора объектов, для которых известен класс или значение свойства объекта. Это можно рассматривать как обучающий набор для алгоритма kNN, хотя никакого явного этапа обучения не требуется. Как в алгоритме классификации, так и в алгоритме регрессии kNN мы можем присвоить вес вкладам соседей. Таким образом, ближайшие соседи вносят больший вклад в средний показатель, чем более отдаленные.

- **Интуиция ближайших соседей**
    
    Интуитивно понятный алгоритм kNN очень прост для понимания. Он просто вычисляет расстояние между точкой выборки данных и всеми другими точками обучающих данных. Расстояние может быть евклидовым или Манхэттенским. Затем он выбирает k ближайших точек данных, где k может быть любым целым числом. Наконец, он присваивает точку выборки данных классу, к которому принадлежит большинство из k точек данных.

   Теперь мы увидим алгоритм kNN в действии. Предположим, у нас есть набор данных с двумя переменными, которые классифицируются как красные и синие.

   В алгоритме kNN k - это число ближайших соседей. Как правило, k - нечетное число, потому что оно помогает определить большинство в классе. Когда k=1, то алгоритм известен как алгоритм ближайшего соседа.

   Теперь мы хотим классифицировать новую точку данных X в синий класс или красный класс. Предположим, что значение k равно 3. Алгоритм kNN начинается с вычисления расстояния между X и всеми остальными точками данных. Затем он находит 3 ближайшие точки с наименьшим расстоянием до точки X.

   На заключительном шаге алгоритма kNN мы присваиваем новую точку данных X большинству из класса 3 ближайших точек. Если 2 из 3 ближайших точек относятся к классу Red, а 1 - к классу Blue, то мы классифицируем новую точку данных как красную.

- **Как определить количество соседей в kNN**

   При построении модели классификатора kNN мне приходит в голову один вопрос: каким должно быть значение ближайших соседей (k), которое дает наибольшую точность. Это очень важный вопрос, потому что точность классификации зависит от нашего выбора k.

   Количество соседей (k) в kNN - это параметр, который нам нужно выбрать во время построения модели. Выбор оптимального значения k в kNN является наиболее важной проблемой. Малое значение k означает, что шум будет оказывать большее влияние на результат. Таким образом, вероятность переобучения очень высока. Большое значение k делает построение модели kNN дорогостоящим с точки зрения вычислений по времени. Кроме того, большое значение k будет иметь более плавную границу принятия решения, что означает меньшую дисперсию, но более высокое смещение.

   Специалисты по обработке данных выбирают нечетное значение k, если количество классов четное. Мы можем применить метод elbow для выбора значения k. Для оптимизации результатов мы можем использовать метод перекрестной проверки. Используя метод перекрестной проверки, мы можем протестировать алгоритм kNN с различными значениями k. Модель, которая обеспечивает хорошую точность, можно считать оптимальным выбором. Это зависит от индивидуальных случаев, и иногда лучшим процессом является прогон каждого возможного значения k и проверка нашего результата.

- **Нетерпеливые ученики против ленивых**

   Нетерпеливые учащиеся имеют в виду, что при предоставлении точек обучающих данных мы построим обобщенную модель, прежде чем выполнять прогнозирование по заданным новым точкам для классификации. Мы можем думать о таких учащихся как о готовых, активных и стремящихся классифицировать новые точки данных.

   Отложенное обучение означает, что нет необходимости в изучении или тренировке модели, и все точки данных используются во время прогнозирования. Ленивые учащиеся ждут до последней минуты, прежде чем классифицировать какую-либо точку данных. Они просто сохраняют обучающий набор данных и ждут, пока не потребуется выполнить классификацию. Ленивые учащиеся также известны как учащиеся на основе экземпляров, потому что ленивые учащиеся хранят точки обучения или экземпляры, и все обучение основано на экземплярах.

   В отличие от нетерпеливых учащихся, ленивые учащиеся выполняют меньше работы на этапе обучения и больше - на этапе тестирования, чтобы составить классификацию.

У нас есть еще один инструмент под названием "Матрица путаницы" - Confusion matrix, который приходит нам на помощь.

При оценке эффективности модели классификации возможны четыре типа результатов. Эти четыре результата описаны ниже:-

Истинно положительные результаты (TP) – Истинно положительные результаты возникают, когда мы предсказываем принадлежность наблюдения к определенному классу, и наблюдение действительно принадлежит этому классу.

Истинные негативы (TN) – Истинные негативы возникают, когда мы предсказываем, что наблюдение не принадлежит к определенному классу, и наблюдение на самом деле не принадлежит к этому классу.

Ложноположительные срабатывания (FP) – Ложноположительные срабатывания возникают, когда мы предсказываем принадлежность наблюдения к определенному классу, но наблюдение на самом деле не принадлежит к этому классу. Этот тип ошибки называется ошибкой I типа.

Ложноотрицательные результаты (FN) – Ложноотрицательные результаты возникают, когда мы предсказываем, что наблюдение не принадлежит к определенному классу, но наблюдение на самом деле принадлежит к этому классу. Это очень серьезная ошибка, и она называется ошибкой второго типа.

Эти четыре результата обобщены в приведенной ниже таблице путаницы.

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
cm_7 = confusion_matrix(y_test, y_pred_7)

print('Confusion matrix\n\n', cm_7)

print('\nTrue Positives(TP) = ', cm_7[0,0])

print('\nTrue Negatives(TN) = ', cm_7[1,1])

print('\nFalse Positives(FP) = ', cm_7[0,1])

print('\nFalse Negatives(FN) = ', cm_7[1,0])
```

Точность может быть определена как процент правильно предсказанных положительных результатов от всех предсказанных положительных результатов. Это может быть задано как отношение истинных срабатываний (TP) к сумме истинных и ложных срабатываний (TP + FP).

Таким образом, точность определяет долю правильно предсказанного положительного результата. Это больше относится к классу положительных, чем к классу отрицательных.

Другим инструментом для визуального измерения производительности классификационной модели является ROC-кривая. ROC-кривая означает кривую рабочих характеристик приемника. Кривая ROC - это график, который показывает эффективность модели классификации на различных пороговых уровнях классификации.

Кривая ROC отображает истинную положительную частоту (TPR) и ложноположительную частоту (FPR) при различных пороговых уровнях.

На кривой ROC мы сосредоточимся на TPR (истинно положительном показателе) и FPR (ложноположительном показателе) для одной точки. Это даст нам общую характеристику кривой ROC, которая состоит из TPR и FPR на различных пороговых уровнях. Итак, кривая ROC отображает соотношение TPR и FPR на разных пороговых уровнях классификации. Если мы снизим пороговые уровни, это может привести к тому, что большее количество товаров будет классифицировано как положительные. Это увеличит как истинные срабатывания (TP), так и ложные срабатывания (FP).

**Строим модель**

Определяем функцию и таргет, делим выборку на обучающие и тестовые данные.

```python
X = data['Message']
y = data['Category_enc']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=34, stratify=y)
```

Теперь мы можем получить tf-idf с помощью TfidfVectorizer от scikit-learn. Tf-idf расшифровывается как частота термина - обратная частоте документа. Он обычно используется при обработке естественного языка для определения наиболее важных слов в документе.

```python
vectorizer = TfidfVectorizer(strip_accents='ascii')

# First fit the vectorizer with our training set
tfidf_train = vectorizer.fit_transform(X_train)

# Now we can fit our test data with the same vectorizer
tfidf_test = vectorizer.transform(X_test)
```

Строим модель:

```python
# Initialize the Multinomial Naive Bayes classifier
nb = MultinomialNB()

# Fit the model
nb.fit(tfidf_train, y_train)

# Print the accuracy score
print("Accuracy:",nb.score(tfidf_test, y_test))
```

**Оцениваем модель**

Показателя точности недостаточно, чтобы сказать, что производительность нашей модели отличная. Нам нужно провести дополнительные расчеты, чтобы быть уверенными.

```python
# Predict the labels
y_pred = nb.predict(tfidf_test)

# Print the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix\n")
print(cm)

# Print the Classification Report
cr = classification_report(y_test, y_pred)
print("\n\nClassification Report\n")
print(cr)

# Print the Receiver operating characteristic Auc score
auc_score = roc_auc_score(y_test, y_pred)
print("\nROC AUC Score:",auc_score)

# Get probabilities.
y_pred_proba = nb.predict(tfidf_test)

# Get False Positive rate, True Positive rate and the threshold
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Visualize the ROC curve.
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('FP Rate')
plt.ylabel('TP Rate')
plt.title('ROC')
plt.show()
```

Область под кривой ROC дала нам меньший результат, но все равно выглядит хорошо.
